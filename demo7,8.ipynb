```python
##total no. of people responded to survey
responded = df2['Country'].value_counts()
responded 
```

```python
##total no. of people with langauage python
countryUsesPython = countryGroup['LanguageAdmired'].apply(lambda x: x.str.contains('Python').sum())
countryUsesPython 
```

```python
##combining two or more series object
df3 = pd.concat([responded,countryUsesPython],axis='columns')
df3
```

```python
df3.rename(columns={'count':'NumResponded','LanguageAdmired':'KnowsPython'},inplace=True)
df3
```

```python
df3['Percentage'] = (df3['KnowsPython']/df3['NumResponded'])*100
```

```python
df3
```

```python
df3.sort_values(by="Percentage", ascending=False, inplace=True)
df3.head(50)
```
